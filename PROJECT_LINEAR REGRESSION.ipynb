```r
    # THE INITIAL HISTOGRAM WITHOUT TRANSFORMATION.

par(mfrow=c(3,3)) # Plotting area setup

# Histogram function
histogram <- function(x) {
  title <- paste(deparse(substitute(x), 500), collapse="\n")
  sdx <- sd(x, na.rm = TRUE)  # Adjust for NA values
  mx <- mean(x, na.rm = TRUE)  # Adjust for NA values
  if(!is.na(sdx) && sdx != 0) {  # Check if standard deviation is not NA or zero
    hist(x, prob=TRUE,
         main=paste("Histogram of ", title),
         xlim=c(mx-3*sdx, mx+3*sdx), ylim=c(0, 0.5/sdx))
    curve(dnorm(x, mean=mx, sd=sdx), col='red', lwd=3, add=TRUE)
  } else {
    cat(paste("Cannot plot", title, "- standard deviation is NA or 0.\n"))
  }
}

         # Histogram
histogram(microwage$age)       # Histogram for 'age'
histogram(microwage$edyrs)     # Histogram for 'edyrs'
histogram(microwage$perwt)     # Histogram for 'perwt'
histogram(microwage$expyrs)    # Histogram for 'expyrs'
histogram(microwage$wkwage)    # Histogram for 'wkwage'
```

```r
        # TRANSFORMATION OF THE ABOVE HISTOGRAM

par(mfrow=c(3,3)) # plotting area setup

# Histogram function that includes the plotting of a normal distribution curve
histogram <- function(x, var_name) {
  sdx <- sd(x, na.rm = TRUE)
  mx <- mean(x, na.rm = TRUE)

  hist(x, prob=TRUE, main=var_name, xlim=c(mx-3*sdx, mx+3*sdx), ylim=c(0, 0.5/sdx), xlab=var_name)
  curve(dnorm(x, mean=mx, sd=sdx), col='red', lwd=2, add=TRUE)  # Add a normal distribution curve
}
           # Histogram Transformation to each variable

histogram(sqrt(microwage$age), "Square Root Transformed Age")      # Square root transformation for 'age'
histogram(sqrt(microwage$edyrs), "Square Root Transformed Edyrs")    # Square root transformation for 'edyrs'
histogram(log(microwage$perwt + 0.05), "Log Transformed Perwt")    # Log transformation for 'perwt'
histogram(sqrt(microwage$expyrs), "Square Root Transformed Expyrs")  # Square root transformation for 'expyrs'
histogram(log(microwage$wkwage + 0.05), "Log Transformed Wkwage")   # Log transformation for 'wkwage'
```

```r
     # FINDING OUTLIERS
# Function to Find Outliers
find_outliers <- function(data, var) {
  variable_data <- data[[var]]
  Q1 <- quantile(variable_data, 0.25, na.rm = TRUE)
  Q3 <- quantile(variable_data, 0.75, na.rm = TRUE)
  IQR <- Q3 - Q1
  lower_bound <- Q1 - 1.5 * IQR
  upper_bound <- Q3 + 1.5 * IQR
  outliers <- variable_data[variable_data < lower_bound | variable_data > upper_bound]
  cat(paste("Outliers for", var, ":", length(outliers), "\n"))
}
      # Find outliers for each variable
find_outliers(microwage, "age")    # Finding outliers in 'age'
find_outliers(microwage, "edyrs")  # Finding outliers in 'edyrs'
find_outliers(microwage, "perwt")  # Finding outliers in 'perwt'
find_outliers(microwage, "expyrs") # Finding outliers in 'expyrs'
find_outliers(microwage, "wkwage") # Finding outliers in 'wkwage'
```

```r
       # REMOVE OUTLIERS
 # Function to Remove Outliers
rm_out <- function(df, v) {
  Q1 <- quantile(df[[v]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df[[v]], 0.75, na.rm = TRUE)
  IQR <- Q3 - Q1
  lb <- Q1 - 1.5 * IQR
  ub <- Q3 + 1.5 * IQR
  df[df[[v]] >= lb & df[[v]] <= ub, ]
}

# Variables list
vars <- c("age", "edyrs", "perwt", "expyrs", "wkwage")

# Removing outliers from all specified variables and storing in new dataset
microwage <- microwage
for(v in vars) {
 microwage <- rm_out(microwage, v)
}

# Dimensions of the original and cleaned data frames
cat("Original data dimensions:", dim(microwage))

par(mfrow=c(2,3)) # Plotting area setup

       # BOX-PLOT
boxplot(microwage$age, main="Box-Plot of Age", xlab="Age")  # Box-Plot for 'age'
boxplot(microwage$edyrs, main="Box-Plot of Edyrs", xlab="Education Years")  # Box-Plot for 'edyrs'
boxplot(microwage$perwt, main="Box-Plot of Perwt", xlab="Person Weight")  # Box-Plot for 'perwt'
boxplot(microwage$expyrs, main="Box-Plot of Expyrs", xlab="Experience Years")  # Box-Plot for 'expyrs'
boxplot(microwage$wkwage, main="Box-Plot of Wkwage", xlab="Weekly Wage")  # Box-Plot for 'wkwage'
```
